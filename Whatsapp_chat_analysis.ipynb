```python
import re
import pandas as pd
```

```python
f=open("WhatsApp Chat with ERP Course.txt","r",encoding="utf-8")
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}\/\d{2,4}\/\d{2,4},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s'
```

```python
massage=re.split(pattern,data)[1:]
massage
```

```python
# \d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'      for am pm format
# df['message_date'].str.replace('pm', 'PM').str.replace('am', 'AM')     you have to replace it
# df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M %p - ')  
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df=pd.DataFrame({"user_massage":massage,"massage_dates":dates})
df['massage_dates'].str.replace('pm', 'PM').str.replace('am', 'AM')    

df["massage_dates"]=pd.to_datetime(df["massage_dates"],format='%d/%m/%Y, %I:%M %p - ')
df.rename(columns={"massage_dates":"date"},inplace=True)
df.head()
```

```python
#separate users and messages
users=[]
messages=[]
for message in df["user_massage"]:
    entry=re.split("([\w\W]+?):\s",message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("group_notification")
        messages.append(entry[0])
df["user"]=users
df["message"]=messages
df.drop(columns=["user_massage"],inplace=True)
df.head()
```

```python
emojis=[]
for message in df["message"]:
    emojis.extend([c for c in message if c in emoji.UNICODE_EMOJI["en"]])
```

```python
words=[]
for messages in df['message']:
    words.extend(messages.split())

#now we have a list which contains all the words in df['message'] of dataframe
#now the goal is to convert it into a string

mystr=""
for x in words:
    mystr=mystr + " "+ x
mystr

#now we have to put this string into the emoji function

myemoji=emoji.emoji_list(mystr)

#myemoji is a list with the following type of elements
#[{'match_start': 5712, 'match_end': 5714, 'emoji': 'üëçüèª'},
 #{'match_start': 6896, 'match_end': 6897, 'emoji': 'üëÜ'},
 #{'match_start': 6897, 'match_end': 6898, 'emoji': 'üëÜ'},
 #{'match_start': 8097, 'match_end': 8098, 'emoji': 'üëÜ'}]

#now we have to extract each emoji from this type of list

pre_final_emoji_list=[]
for i in range(len(myemoji)):
    pre_final_emoji_list.extend(myemoji[i]['emoji'])

#now we have another list which contains all the emojis but it also contains some extra emojis which need to be removed
#we create two new list, one where removal characters are present and another one where final emojis are.

emojis_to_be_removed=['','üèª']

final_emoji_list=[]
for items in pre_final_emoji_list:
    if items not in emojis_to_be_removed:
        final_emoji_list.append(items)
final_emoji_list
```

```python
pd.DataFrame(Counter(final_emoji_list).most_common(len(Counter(final_emoji_list))))
```
